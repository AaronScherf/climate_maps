from urllib.request import urlopen
from zipfile import ZipFile

    # Download the file from the URL

    # Create a new file on the hard drive
tempzip = open("ca-state-boundary/ca-state-boundary.zip", "wb")
    # Write the contents of the downloaded file into the new file
tempzip.write(zipresp.read())
    # Close the newly-created file
tempzip.close()
    # Re-open the newly-created file with ZipFile()
zf = ZipFile("ca-state-boundary/ca-state-boundary.zip")
    # Extract its contents into <extraction_path>
    # note that extractall will automatically create the path
zf.extractall(path = 'ca-state-boundary')
    # close the ZipFile instance
zf.close()

## Import Feature Collection Object from GEE


```python
ca_bound = ee.FeatureCollection('users/theaaronscherf/ca-state-boundary')
ca_bound.getInfo()
```

```python
cropLandcover = ee.ImageCollection('USDA/NASS/CDL') \
                  .filter(ee.Filter.date('2018-01-01', '2019-12-31')) \
                  .first().select('cropland')
Map.addLayer(ca_bound, {}, 'CA')
Map.addLayer(cropLandcover.clip(ca_bound), {}, 'Crop Landcover Class')
Map
```

```python
EVI = ee.ImageCollection("NOAA/VIIRS/001/VNP13A1") \
    .filterDate('2018-01-01', '2018-01-31') \
    .select('EVI') \
    .mean()

evi_vis = {
  'min': 0.0,
  'max': 10000.0,
  'palette': ["000", "040", "080", "0B0", "0F0"],
}

Map.addLayer(EVI.clip(ca_bound), evi_vis, 'Mean EVI January 2018')
Map
```

```python
naip_ts = geemap.naip_timeseries(start_year=2009, end_year=2018)
```

```python
layer_names = ['NAIP ' + str(year) for year in range(2009, 2019)]
print(layer_names)
```

```python
naip_vis = {'bands': ['N', 'R', 'G']}
```

```python
Map = geemap.Map()
Map.ts_inspector(left_ts=naip_ts, right_ts=naip_ts, left_names=layer_names, right_names=layer_names, left_vis=naip_vis, right_vis=naip_vis)
Map
```
